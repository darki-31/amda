# Data calculation

Before we can plot something we have to calculate the data for position, distance travelled and spatial preference.

Import and init ActivityMonitorDataAnalyzer.
- time_frame_ms is time frame for position calculation in milliseconds.
- min_pos_dist is minimal distance between the new and old mean position to be acepted. This can be seen as a filter that counteracts the random jitter of the mean position. Especially when the animal is not really moving.

```python
from amda import ActivityMonitorDataAnalyzer
an = ActivityMonitorDataAnalyzer(time_frame_ms=500, min_pos_dist=120)
```

Now we have two options:
1. calculate all data with an.calculate("file_name"). This may take a while, depending on the amount of data.
2. scan a folder for calculated files with an.scan_folder("folder_name"). Choose this option if you already calculated your data.

```python
an.calculate("data/2017_05_31/Sample-17.05.31-Test_SAM_FTP_YW_KF.csv")
#an.scan_folder("data/2017_05_31")
```

In both cases we can now check files available for plotting.

```python
print(an.pref_files)
print(an.distance_files)
```

# Plot roaming entropie

## Plot all data in one chart

```python
an.plot_roaming_entropie(period="1H", fig_size=(23,6))
```

## Plot each day in one chart

Over the "period" parameter you can change the time window in which the roaming entropy is calculated. Following codes are valid:
1. "1H" - one hour
2. "30T" - 30 minutes
3. "15T" - 15 minutes
4. "10T" - 10 minutes
5. "5T" - 5 minutes

```python
an.plot_roaming_entropie_per_day(period="1H", fig_size=(23,12))
```
